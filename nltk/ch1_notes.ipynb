## Language Processing and Python

```python
import nltk
nltk.download();
```

```python
import nltk.book
```

#### Searching text

```python
nltk.book.text3.concordance("lived")
```

```python
nltk.book.text1.similar("monstrous")
```

```python
nltk.book.text2.similar("monstrous")
```

**Common contexts** are the contexts shared by two words.  
The "border" words that are shared by the input words is returned:

```python
nltk.book.text3.common_contexts(["Adam", "Seth"])
```

This means the text "and Adam lived" and "and Seth lived" both appear in the text, as can be seen above.

```python
nltk.book.text4.dispersion_plot(["citizens", "democracy", "republic", "freedom", "liberty", "economy", "labor", "jobs"])
```

The generate method builds and ngram index and generates text using a trigram language model:

```python
nltk.book.text3.generate()
```

#### Simple Statistics

```python
moby_dick_fdist = nltk.book.FreqDist(nltk.book.text1)
print(moby_dick_fdist)
```

```python
moby_dick_fdist.most_common(20)
```

```python
moby_dick_fdist['whale']
```

```python
moby_dick_fdist.plot(50, cumulative=True)
```

**Hapaxes** are words that only occur once:

```python
moby_dick_fdist.hapaxes()[:20]
```

```python
chats_fdist = nltk.book.FreqDist(nltk.book.text5)
sorted(w for w in set(nltk.book.text5) if len(w) > 7 and chats_fdist[w] > 7)
```

```python
genesis_fdist = nltk.book.FreqDist(nltk.book.text3)
sorted(w for w in set(nltk.book.text3) if len(w) > 7 and genesis_fdist[w] > 7)
```

A **bigram** is a sequential word pair:

```python
list(nltk.book.bigrams(['more', 'is', 'said', 'than', 'done']))
```

A **collocation** is a sequence of words that occur together unusually often---like 'red wine' but not 'the wine'.  
They are bigrams that appear more often than usual given the freqency of their individual words:

```python
nltk.book.text4.collocations()
```

```python
nltk.book.text8.collocations()
```

Getting word lengths:

```python
len_fdist = nltk.book.FreqDist(len(w) for w in nltk.book.text1)
print(len_fdist)
len_fdist
```

```python
len_fdist.most_common()
```

```python
len_fdist[len_fdist.max()]
```

```python
len_fdist.freq(3)
```

```python
len_fdist.tabulate()
```

```python
len_fdist.N()
```

```python
len(len_fdist)
```

Update one fdist with another's data:

```python
len_fdist |= genesis_fdist
len_fdist.tabulate(25)
```
