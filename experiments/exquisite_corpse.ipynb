## Exquisite Corpse
#### Experiments with text and randomness

#### What if Jane Austen wrote Leaves of Grass?

```python
import nltk
import spacy
from nltk.corpus import gutenberg

gutenberg.fileids()
```

```python
nlp = spacy.load("en_core_web_sm")
emma_doc = nlp(gutenberg.raw('austen-emma.txt'))
```

Build a unigram language model from 'Emma' text:

```python
import collections

emma_unigram = collections.defaultdict(list)
for token in emma_doc:
    emma_unigram[token.pos_, token.dep_].append(token.text)
print(emma_unigram['DET', 'det'][:10])
```

Get parts-of-speech from 'Leaves of Grass' and use as a template:

```python
import random

whitman_doc = nlp(gutenberg.raw('whitman-leaves.txt'))
austen_leaves_of_grass = []
for token in whitman_doc:
    if token.pos_ == 'PUNCT':
        austen_leaves_of_grass.append(token.text)
    else:
        if emma_unigram[token.pos_, token.dep_]:
            rand_pos_word = random.choice(emma_unigram[token.pos_, token.dep_])
        austen_leaves_of_grass.append(rand_pos_word)
```

```python
print(' '.join(austen_leaves_of_grass))
```
